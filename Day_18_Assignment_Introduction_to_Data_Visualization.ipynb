# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
monthly_sales = df.groupby('month_number')[['facecream', 'facewash']].sum()


monthly_sales.plot(kind='bar', width=0.7, color=['#1f77b4', '#ff7f0e'])


plt.xlabel('Month Number', fontsize=12, labelpad=10)
plt.ylabel('Units Sold', fontsize=12, labelpad=10)
plt.title('Face Cream vs. Face Wash Units Sold Per Month (Grouped)', fontsize=14, fontweight='bold', pad=15)




```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here

import pandas as pd
import matplotlib.pyplot as plt

product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
monthly_sales = df.groupby('month_number')[product_columns].sum()

monthly_sales.plot(kind='line', marker='o', linewidth=2, markersize=5)

plt.xlabel('Month Number', fontsize=12, labelpad=10)
plt.ylabel('Units Sold', fontsize=12, labelpad=10)
plt.title('Product Units Sold Per Month (Multiline Trend)', fontsize=14, fontweight='bold', pad=15)

```
