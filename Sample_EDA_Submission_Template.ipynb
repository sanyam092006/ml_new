#### Chart - 1

````python
# Chart - 1 visualization code
on_time = ((df['DEP_DELAY'] <= 0) & (df['CANCELLED'] == 0) & (df['DIVERTED'] == 0)).sum()
delayed = ((df['DEP_DELAY'] > 0) & (df['CANCELLED'] == 0) & (df['DIVERTED'] == 0)).sum()
cancelled = (df['CANCELLED'] == 1).sum()
diverted = (df['DIVERTED'] == 1).sum()

status = pd.DataFrame({
    'Flight Status': ['On Time', 'Delayed', 'Cancelled', 'Diverted'],
    'Count': [on_time, delayed, cancelled, diverted]
})


plt.figure(figsize=(8,5))
plt.bar(status['Flight Status'], status['Count'])
plt.title('Overall Flight Status Distribution')
plt.xlabel('Flight Status')
plt.ylabel('Number of Flights')


for i, value in enumerate(status['Count']):
    plt.text(i, value, f'{value:,}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
````

From a business perspective, flight delays increase operating expenses through additional fuel consumption, crew overtime, aircraft repositioning, airport charges, and passenger compensation. Using common industry estimates of approximately 80 dollar per  delay  minute and around 30,000 dollar per cancelled flight, the combined financial impact of delays and cancellations in a dataset of this size could exceed 1 billion dollar in operational costs.

**Negative Growth Insight:-**

The high number of delayed flights represents the greatest threat to business growth. Persistent delays reduce customer confidence, increase compensation and operating costs, and can lead to lower customer retention. Similarly, flight cancellations result in lost revenue, rebooking expenses, and reputational damage. If these operational issues are not addressed, they can significantly affect an airline's financial performance and long-term competitiveness.

#### Chart - 2

````python
import matplotlib.pyplot as plt


airline_delay = (
    df[df["CANCELLED"]==0]
    .groupby('AIRLINE')['DEP_DELAY']
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))
airline_delay.plot(kind='bar')

plt.title('Average Departure Delay by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Delay (Minutes)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
````

##### 1. Why did you pick the specific chart?

Because This Directly Answers One Of The Most Important Questions Which is Which Airline Is Responsible For Highest Delays?

##### 2. What is/are the insight(s) found from the chart?

Answer Here

*   JetBlue Airways has the highest average departure delay.
*   Alaska Airlines demonstrates the best departure performance.
*   Legacy and regional carriers generally perform better.
*   There is a considerable variation in departure performance across airlines.

#### Chart - 3

````python
# Chart - 3 visualization code
import matplotlib.pyplot as plt

# Sum delay minutes for each delay cause
delay_causes = df[
    [
        'DELAY_DUE_CARRIER',
        'DELAY_DUE_WEATHER',
        'DELAY_DUE_NAS',
        'DELAY_DUE_SECURITY',
        'DELAY_DUE_LATE_AIRCRAFT'
    ]
].fillna(0).sum()


delay_causes.index = [
    'Carrier',
    'Weather',
    'NAS',
    'Security',
    'Late Aircraft'
]


plt.figure(figsize=(10,6))
delay_causes.plot(kind='bar')

plt.title('Total Delay Minutes by Delay Cause')
plt.xlabel('Delay Cause')
plt.ylabel('Total Delay Minutes')

plt.tight_layout()

plt.show()
````

##### 1. Why did you pick the specific chart?

This Chart Identifies The Root Cause Of Those Delays

Answer Here

*   Late Aircraft is the largest contributor to flight delays.
*   Carrier-related issues are the second-largest cause of delays.
*   NAS (National Airspace System) delays have a noticeable impact.
*   Weather contributes relatively fewer delay minutes.
