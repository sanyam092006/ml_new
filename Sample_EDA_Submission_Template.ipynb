#### Chart - 1

````python
# Chart - 1 visualization code
on_time = ((df['DEP_DELAY'] <= 0) & (df['CANCELLED'] == 0) & (df['DIVERTED'] == 0)).sum()
delayed = ((df['DEP_DELAY'] > 0) & (df['CANCELLED'] == 0) & (df['DIVERTED'] == 0)).sum()
cancelled = (df['CANCELLED'] == 1).sum()
diverted = (df['DIVERTED'] == 1).sum()

status = pd.DataFrame({
    'Flight Status': ['On Time', 'Delayed', 'Cancelled', 'Diverted'],
    'Count': [on_time, delayed, cancelled, diverted]
})


plt.figure(figsize=(8,5))
plt.bar(status['Flight Status'], status['Count'])
plt.title('Overall Flight Status Distribution')
plt.xlabel('Flight Status')
plt.ylabel('Number of Flights')


for i, value in enumerate(status['Count']):
    plt.text(i, value, f'{value:,}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
````

#### Chart - 2

````python
import matplotlib.pyplot as plt


airline_delay = (
    df[df["CANCELLED"]==0]
    .groupby('AIRLINE')['DEP_DELAY']
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))
airline_delay.plot(kind='bar')

plt.title('Average Departure Delay by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Delay (Minutes)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
````

#### Chart - 4

````python
import matplotlib.pyplot as plt

# Extract scheduled departure hour
df['DEP_HOUR'] = df['CRS_DEP_TIME'] // 100

# Median departure delay by hour
hourly_delay_median = (
    df[df['CANCELLED'] == 0]
    .groupby('DEP_HOUR')['DEP_DELAY']
    .median()
)

# Plot
plt.figure(figsize=(10,6))
hourly_delay_median.plot(kind='line', marker='o')

plt.title('Median Departure Delay by Scheduled Departure Hour')
plt.xlabel('Scheduled Departure Hour')
plt.ylabel('Median Delay (Minutes)')
plt.xticks(range(24))
plt.grid(True)

plt.tight_layout()
plt.show()
````
